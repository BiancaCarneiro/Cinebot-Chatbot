```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
modelos = ["CNN", "LSTM", "SVM", "Transformers"]
classificador_acc = [0.98, 1, 0.97, 0.9879]
classificador_f1 = [0.98, 1, 0.97, 0.9879]
```

```python
modelos_ner = ["LSTM", "Transformers"]
ner_acc = [0.6, .7928]
ner_f1 = [0.43, .8025]
```

```python
ax = sns.barplot(x=modelos, y=classificador_acc, hue=modelos)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Comparação da acurácia dos modelos de classificação")
plt.savefig("imgs/classificador_acc.png")
plt.tight_layout()
plt.show()
```

```python
ax = sns.barplot(x=modelos, y=classificador_f1, hue=modelos)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Comparação do F1 dos modelos de classificação")
plt.savefig("imgs/classificador_f1.png")
plt.tight_layout()
plt.show()
```

```python
ax = sns.barplot(x=modelos_ner, y=ner_acc, hue=modelos_ner)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Comparação da acurácia dos modelos de extração")
plt.savefig("imgs/ner_acc.png")
plt.tight_layout()
plt.show()
```

```python
ax = sns.barplot(x=modelos_ner, y=ner_f1, hue=modelos_ner)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Comparação do F1 dos modelos de extração")
plt.savefig("imgs/ner_f1.png")
plt.tight_layout()
plt.show()
```
